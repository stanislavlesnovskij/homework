```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df=pd.read_csv('2019.csv',sep=',')
df.info()
```

```python
df.describe()
```

```python
plt.plot(df['Generosity'],df['Country or region'])
plt.title('Зависимость уровня щедрости от страны')
plt.xlabel('Уровень щедрости')
plt.ylabel('Страна')
#Можно сделать вывод, что уровень щедрости не зависит от развитости страны и не влияет на уровень счастья.
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 20,6
```

```python
df.style.bar(align='mid')
#Можно сделать вывод, что уровень счастья зависит от качества социальной поддержки,свободы выбора и продолжительности жизни.
```

```python
df_new = df[['GDP per capita','Country or region','Perceptions of corruption']].sort_values(by='Perceptions of corruption',ascending=False)
```

```python
df_pie=df_new[['GDP per capita','Country or region']].head(5)
```

```python
pie=df_pie['GDP per capita'].plot(kind='pie')
plt.title('ВВП стран с наибольшими представлениями о коррупции')
plt.labels=df_pie['Country or region']
#Можно сделать вывод, что в целом ВВП и представления о коррупции мало связаны
```

```python
df_sort=df.sort_values(by='Score')
sns.regplot(data=df_sort,x='Healthy life expectancy',y='Country or region',color='red',marker='^')
plt.title('Уровень ожидаемой продолжительности жтзни в зависиости от страны')
#Прослеживается некоторая зависимость между продолжительностью жизни и уровнем счастья
```
