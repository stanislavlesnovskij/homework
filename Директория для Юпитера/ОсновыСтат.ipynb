```python
horse_health1=df[['surgery','age','rectal_temp','pulse','resp_rate','temperature of extremities','pain','outcome']]
```

```python
horse_health=horse_health1.replace('?','0')
horse_health.head()
```

```python
column=['surgery','age','rectal_temp','pulse','resp_rate','temperature of extremities','pain','outcome']
for el in column:
    desc=horse_health[el].astype('float64').describe()
    print(desc)
```

```python
#Основные метрики
column=['surgery','age','rectal_temp','pulse','resp_rate','temperature of extremities','pain','outcome']
for el in column:
    horse_health1[el].astype('float64').fillna(horse_health[el].astype('float64').median())
    min_=horse_health[el].min()
    max_=horse_health[el].max()
   
    mean_=horse_health[el].astype('float64').mean()
   
      
    mode_=horse_health[el].astype('float64').round().mode()[0]
    median_=horse_health[el].astype('float64').median()
    std_=horse_health[el].astype('float64').std()
    var_=horse_health[el].astype('float64').var()
    q1=horse_health[el].astype('float64').quantile(0.25)
    q3=horse_health[el].astype('float64').quantile(0.75)
    Lower_outlier =q1-(1.5*(q3-q1))
    Higher_outlier = q3+(1.5*(q3-q1))
    print('min=',min_)
    print('max=',max_)
    print('mean=',mean_)
    print('mode=',mode_)
    print('med=',median_)
    print('std=',std_)
    print('var=',var_) 
    print('q1=',q1)
    print('q3=',q3)
    print('---------------------')
```
