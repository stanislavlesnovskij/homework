```python
from scipy.stats import t
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots(1, 1)
df = 2.74
df1=6
df2=10
df3=200
df4=30
meanT, varT, skewT, kurtT = t.stats(df, moments='mvsk')
meanT1, varT1, skewT1, kurtT1 = t.stats(df1, moments='mvsk')
meanT2, varT2, skewT2, kurtT2 = t.stats(df2, moments='mvsk')
meanT3, varT3, skewT3, kurtT3 = t.stats(df3, moments='mvsk')
meanT4, varT4, skewT4, kurtT4 = t.stats(df4, moments='mvsk')


xT = np.linspace(t.ppf(0.01, df),
                t.ppf(0.99, df), 100)
xT1 = np.linspace(t.ppf(0.01, df1),
                t.ppf(0.99, df1), 100)
xT2 = np.linspace(t.ppf(0.01, df2),
                t.ppf(0.99, df2), 100)
xT3 = np.linspace(t.ppf(0.01, df3),
                t.ppf(0.99, df3), 50)
xT4 = np.linspace(t.ppf(0.01, df4),
                t.ppf(0.99, df4), 100)
ax.plot(xT, t.pdf(xT, df),
       'r-', lw=5, alpha=0.6, label='t pdf')
ax.plot(xT1, t.pdf(xT1, df1),
       'g-', lw=5, alpha=0.6, label='t pdf')
ax.plot(xT2, t.pdf(xT2, df2),
       'b-', lw=5, alpha=0.6, label='t pdf')
ax.plot(xT3, t.pdf(xT3, df3),
       'y-', lw=5, alpha=0.6, label='t pdf')
ax.plot(xT4, t.pdf(xT4, df4),
       'm-', lw=5, alpha=0.6, label='t pdf')
```

```python
from scipy.stats import expon
fig, ax = plt.subplots(1, 1)
meanE, varE, skewE, kurtE = expon.stats(df, moments='mvsk')
meanE1, varE1, skewE1, kurtE1 = expon.stats(df1, moments='mvsk')
meanE2, varE2, skewE2, kurtE2 = expon.stats(df2, moments='mvsk')
meanE3, varE3, skewE3, kurtE3 = expon.stats(df3, moments='mvsk')
meanE4, varE4, skewE4, kurtE4 = expon.stats(df4, moments='mvsk')
xE = np.linspace(expon.ppf(0.1),
                expon.ppf(0.5), 50)
xE1 = np.linspace(expon.ppf(0.01),
                expon.ppf(0.2), 200)
xE2 = np.linspace(expon.ppf(0.01),
                expon.ppf(0.4), 150)
xE3 = np.linspace(expon.ppf(0.01),
                expon.ppf(0.6), 300)
xE4 = np.linspace(expon.ppf(0.01),
                expon.ppf(0.8), 250)
ax.plot(xE, expon.pdf(xE),
       'r-', lw=5, alpha=0.6, label='expon pdf')
ax.plot(xE1, expon.pdf(xE1),
       'g-', lw=5, alpha=0.6, label='expon pdf')
ax.plot(xE2, expon.pdf(xE2),
       'y-', lw=5, alpha=0.6, label='expon pdf')
ax.plot(xE3, expon.pdf(xE3),
       'm-', lw=5, alpha=0.6, label='expon pdf')
ax.plot(xE4, expon.pdf(xE4),
       's-', lw=5, alpha=0.6, label='expon pdf')
```
