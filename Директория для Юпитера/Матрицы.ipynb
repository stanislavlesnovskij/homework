```python
import numpy as np
import matplotlib.pyplot as plt

A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)
x = A[:,0]
y = A[:,1]

```

```python
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
B = A*0.5
```

```python
B = np.array(B)
x = B[:,0]
y = B[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
M=np.array([[np.cos(130), np.sin(130), 0],[-np.sin(130), np.cos(130), 0],[0, 0, 1]])
```

```python
G=np.dot(B,M)
G
```

```python
G = np.array(G)
x = G[:,0]
y = G[:,1]
plt.plot(x,y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
O=np.array([[0, 1, 0], [1, 0, 0], [0, 0, 1]])
```

```python
F=np.dot(G,O)
F
```

```python
F= np.array(F)
x = F[:,0]
y = F[:,1]
plt.plot(x,y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
F= np.array(F)
x = F[:,0]
y = F[:,1]
x=x+200
y=y+300
plt.plot(x,y)
plt.ylim([-200, 500])
plt.xlim([-200, 500])
```

```python
F= np.array(F)
x = F[:,0]
y = F[:,1]
x=x+200
x=x*0.5
y=y+300
y=y*1.2
plt.plot(x,y)
plt.ylim([0, 500])
plt.xlim([0, 500])
```
